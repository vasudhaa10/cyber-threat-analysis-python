## 📊 Step 5: Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(8,5))

df["Attack Type"].value_counts().plot(
    kind="bar",
    color="steelblue",
    edgecolor="black"
)

plt.title("Distribution of Cyber Attack Types", fontsize=15, fontweight="bold")
plt.xlabel("Attack Type", fontsize=12)
plt.ylabel("Number of Attacks", fontsize=12)

plt.xticks(rotation=30)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

```python
df["Severity Level"].value_counts()
```

```python
plt.figure(figsize=(6,6))

df["Severity Level"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=["#ff9999", "#66b3ff", "#99ff99"]
)

plt.title("Severity Level Distribution", fontsize=15, fontweight="bold")
plt.ylabel("")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df["Protocol"].value_counts().plot(
    kind="barh",
    color="seagreen",
    edgecolor="black"
)

plt.title("Network Protocol Distribution", fontsize=15, fontweight="bold")
plt.xlabel("Number of Records", fontsize=12)
plt.ylabel("Protocol", fontsize=12)

plt.grid(axis="x", linestyle="--", alpha=0.5)

plt.show()
```

```python
df["Network Segment"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Network Segments",fontsize=15, fontweight="bold")
plt.xlabel("Network Segment",fontsize=12)
plt.ylabel("Count",fontsize=12)

plt.show()
```

```python
df["Action Taken"].value_counts().plot(
    kind="bar",
    color="steelblue",
    edgecolor="black"
)

plt.title("Action Taken Distribution",fontsize=15, fontweight="bold")
plt.xlabel("Action Taken",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(7,7))

colors = ["#4E79A7", "#F28E2B", "#59A14F", "#E15759"]

plt.pie(
    df["Traffic Type"].value_counts(),
    labels=df["Traffic Type"].value_counts().index,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    wedgeprops={"width":0.45, "edgecolor":"white"}
)

plt.title("Traffic Type Distribution", fontsize=15, fontweight="bold")

plt.show()
```

```python
import matplotlib.pyplot as plt

df["Anomaly Scores"].plot(kind="hist", bins=20)

plt.title("Anomaly Scores Distribution")
plt.xlabel("Anomaly Score")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Log Source"].value_counts().plot(kind="bar")

plt.title("Log Source")
plt.xlabel("Source")
plt.ylabel("Count")
plt.show()
```

## 🔥 Step 6: Relationship Analysis

```python
attack_severity = pd.crosstab(df["Attack Type"], df["Severity Level"])
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    attack_severity,
    annot=True,
    cmap="YlOrRd",
    fmt="d"
)

plt.title("Attack Type vs Severity Level", fontsize=15, fontweight="bold")

plt.show()
```
